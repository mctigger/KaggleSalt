```python
import matplotlib.pyplot as plt
%matplotlib inline

import datasets
import utils

samples = utils.get_aux_samples()
dataset = datasets.DataAnalysisDataset(samples, 'data/auxiliary_small')
rows = 100
cols = 2
fig, ax = plt.subplots(rows, cols, figsize=(cols * 3, rows * 3))
for i, (image, mask, id) in enumerate(dataset):
    if i >= rows:
        break
        
    ax[i][0].imshow(image)
    ax[i][1].imshow(mask)
    


plt.tight_layout()
```
